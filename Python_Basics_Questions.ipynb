    # question 1: what is python, and why is it popular?
    # answer:
    # python is a high-level, interpreted, and easy-to-learn programming language.
    # it is popular because of its simple syntax, large library support,
    # cross-platform compatibility, and wide use in web development,
    # data science, artificial intelligence, automation, and software development.

    # question 2: what is an interpreter in python?
    # answer:
    # an interpreter is a program that executes python code line by line
    # without converting the entire program into machine code first.
    # it helps in easy debugging and quick program execution.
    
    # question 3: what are pre-defined keywords in python?
    # answer:
    # pre-defined keywords are reserved words that have special meanings
    # in python and cannot be used as variable names.
    # examples: if, else, for, while, class, def, return, import, true, false.
    
    # question 4: can keywords be used as variable names?
    # answer:
    # no, python keywords cannot be used as variable names because
    # they are reserved for specific purposes in the language.
    
    # question 5: what is mutability in python?
    # answer:
    # mutability means an object can be changed after it is created.
    # mutable objects can be modified, while immutable objects cannot.
    # examples:
    # mutable: list, dictionary, set.
    # immutable: tuple, string, integer.
    
    # question 6: why are lists mutable, but tuples are immutable?
    # answer:
    # lists are mutable because their elements can be added, removed,
    # or modified after creation.
    # tuples are immutable because their values cannot be changed
    # once they are created.
    
    # question 7: what is the difference between "==" and "is" operators in python?
    # answer:
    # == compares the values of two objects.
    # is compares whether two variables refer to the same object in memory.
    # example:
    # == checks value equality.
    # is checks object identity.
    
    # question 8: what are logical operators in python?
    # answer:
    # logical operators are used to combine conditional statements.
    # types:
    # and - returns true if both conditions are true.
    # or - returns true if at least one condition is true.
    # not - reverses the result of a condition.
    
    # question 9: what is type casting in python?
    # answer:
    # type casting is the process of converting one data type
    # into another data type.

    # examples:
    # int()
    # float()
    # str()
    # bool()
    
    # question 10: what is the difference between implicit and explicit type casting?
    # answer:
    # implicit type casting is performed automatically by python.

    # explicit type casting is performed manually by the programmer
    # using functions such as int(), float(), str(), and bool().
    
    # question 11: what is the purpose of conditional statements in python?
    # answer:
    # conditional statements are used to make decisions in a program.
    # they execute different blocks of code based on whether
    # a condition is true or false.

    # examples:
    # if
    # if-else
    # if-elif-else
    
    # question 12: how does the elif statement work?
    # answer:
    # the elif statement checks another condition if the previous
    # if or elif condition is false.
    # it allows multiple conditions to be tested one after another.
    
    # question 13: what is the difference between for and while loops?
    # answer:
    # for loop:
    # used when the number of iterations is known.
    # commonly used with sequences like lists and strings.

    # while loop:
    # used when the number of iterations is unknown.
    # it runs until the given condition becomes false.
    
    # question 14: describe a scenario where a while loop is more suitable than a for loop.
    # answer:
    # a while loop is more suitable when the number of iterations
    # is not known in advance.
    # example:
    # asking the user to enter the correct password repeatedly
    # until the correct password is entered.

```python
# question 4: write a program that checks if a given word is a python keyword.

import keyword

word = input("enter a word: ")

if keyword.iskeyword(word):
    print(word, "is a python keyword.")
else:
    print(word, "is not a python keyword.")
```

```python
# question 5: create a list and tuple in python, and demonstrate how
# attempting to change an element works differently for each.

my_list = [10, 20, 30]
my_tuple = (10, 20, 30)

print("original list:", my_list)
my_list[1] = 99
print("modified list:", my_list)

print("tuple:", my_tuple)

try:
    my_tuple[1] = 99
except TypeError as e:
    print("error:", e)
```

```python
# question 6: write a function to demonstrate the behavior of
# mutable and immutable arguments.

def modify(num, lst):
    num += 5
    lst.append(50)

x = 10
y = [10, 20]

modify(x, y)

print("integer:", x)
print("list:", y)
```

```python
# question 7: write a program that performs basic arithmetic
# operations on two user-input numbers.

a = float(input("enter first number: "))
b = float(input("enter second number: "))

print("addition:", a + b)
print("subtraction:", a - b)
print("multiplication:", a * b)
print("division:", a / b)
```

```python
# question 10: write code to demonstrate type casting with list elements.

numbers = ["10", "20", "30", "40"]

int_numbers = list(map(int, numbers))

print("original list:", numbers)
print("converted list:", int_numbers)
```

```python
# question 11: write a program that checks if a number is
# positive, negative, or zero.

num = float(input("enter a number: "))

if num > 0:
    print("positive")
elif num < 0:
    print("negative")
else:
    print("zero")
```

```python
# question 12: write a for loop to print numbers from 1 to 10.

for i in range(1, 11):
    print(i)
```

```python
# question 13: write a python program to find the sum
# of all even numbers between 1 and 50.

total = 0

for i in range(2, 51, 2):
    total += i

print("sum of even numbers:", total)
```

```python
# question 14: write a program to reverse a string using a while loop.

text = input("enter a string: ")

reverse = ""
i = len(text) - 1

while i >= 0:
    reverse += text[i]
    i -= 1

print("reversed string:", reverse)
```

```python
# question 15: write a python program to calculate the factorial
# of a number using a while loop.

num = int(input("enter a number: "))

fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("factorial:", fact)
```
